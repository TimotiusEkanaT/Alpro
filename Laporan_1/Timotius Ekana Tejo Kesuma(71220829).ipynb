```python
#no 1

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10)
y = np.sin(x)
z= np.cos(x)
plt.plot(x,y, 'b', x,z,'r')
plt.xlabel('Radians'); plt.ylabel('value');
plt.title('Plotting Demonstration')
plt.legend(['Sin', 'Cos'])
plt.grid()
```
